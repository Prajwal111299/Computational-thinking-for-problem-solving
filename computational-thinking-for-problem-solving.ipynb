**COMPUTAIONAL THINKING FOR PROBLEM SOLVING -**

4 pillars for problem solving ->

**DECOMPOSITION** : Breaking a complex problem into more manageable sub-problems. Putting the solutions to the sub-problems together gives a solution to the original, complex problem.
Eg- Stiching the images of Europe from satellite, then zooming in or out.

**PATTERN RECOGNITION** : Finding similarities or shared characteristics within or between problems. Makes the problem easier to solve since the same solution can be used for each occurence of the pattern.
Eg- Compressing data, social media site : Photo Albums.

**DATA REPRESENTATIONS & ABSTRACTION** : Determining what characteristics of the problem are important and filtering out those that are not. Use those to create a representation of what we are trying to solve.
Eg- student_id, Online retail Website(Book names).

**ALGORITHMS** : Step-by-step instruction of how to solve a problem. Identifies what is to be done(the instructions) and the order in which they should be done.
Eg- Making o cup of tea ->
-> Fill electric tea Kettle
-> Bring it to a bail
-> Put teabag in cup
-> Steep for 4 minutes
-> Remove teabag
Can be described in English, flowchart or pseudocode.

**WHAT IS AN INSTRUCTION?**
Often express as something human understanf. Eventually translates to a code for computers.

**LINEAR SEARCH** : Searching for an element in an unordered list follows some of the same patterns as other algorithms we've seen before. In some cases we need to look through the entire list, but in other cases we don't.

**ALGORITHM COMPLEXITY** :


**Binary search has logarithmic complexity : if we double the input size, the number of operations only increases by one.**

**Brute force algorithm** : Brute force algorithms can be used for solving optimization problems ->
-> Try all possible solution.
-> Identify the ones that solve the problem.
-> Determine the one that is best.
there can be many possible solutions!
For an input of n elements -> Exponential : 2n
                              Factorial : n!
                              
**GREEDY ALGORITHMS** : Do the thing that seems best at the time. Brute force : consider all possible solutions and then choose the one that is best. Greedy : make a decision that seems best right now and keep repeating until done.
can be used for solving optimization problems.




**A HISTORY OF COMPUTER** ->

**Charles Babbage** : Created the first concept of a programmable computer. Difference Engine (1822) to calculate mathematical tables. Analytical Engine (1834) as a general-purpose steam-engine powered machine with memory storage, a central processing unit, and printer.

**Ada Lovelace** : First computer programmer. Foresaw that Babbage's Analytical purpose could serve a more general purpose, Eg-> create graphics or compose music.

**George Boole** : Established modern symbolic logic

**Executing an Instruction** :
1. Fetch - Read the instrution from memory.
2. Decode - Figure out what it means and put input(operands) in right place.
3. Execute - Perform the instruction and put output(results) in right place.
4. Update - set the address of next instruction to execute.
